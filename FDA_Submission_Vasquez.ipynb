## Algorithm Description 

### 1. General Information

**Intended Use Statement:** This algorithm is intended for assisting a radiologist in pneumonia identification from chest x-rays.

**Indications for Use:**  It is indicated for use in male and female patients within the age range from 10-89 years with chest x-rays taken in the AP and PA view positions on a ER setting. The algorithm should be integrated into the normal workflow of the diagnostic clinics. The X-Ray images should be available in DICOM format. About the workflow, the data is sent through the algorithm and it checks the required conditions. If the images satisfies the criteria, the algorithm makes the prediction. Once the prediction is complete, this data is sent to an expert radiologist/doctor, and then he/she will give the final verdict based on his/her independent diagnosis as well as the data supplied by the algorithm.

**Device Limitations:** Acording to the population used for training the algorithm. Patient ages are limited to above 10 and below 89 given that the majority of examples are between these ages. The algorithm is limited to Male or Female genders according to the dataset labels. The inference time reported is based on machines with GPU available, the lack of GPU may lead to procesing delays but the accuracy is not compromised. This algoithm's outputs must always be verified by a expert doctor or radiologist.

**Clinical Impact of Performance:** Algorithm performance shows that it will be useful for screening chest x-rays for pneumonia and may also be used for workflow prioritization. Errors in the algorithm predictions may lead to severe consequences compromising life. False negative predictions might be a life threat, since important treatments could be omitted. False positives could also have important phisical and psicological consequences on the patient.

### 4. Databases

** Dataset subsampling **
A reduced dataset composed of 2862 examples was created, 1431 for pneumonia and 1431 for no neumonia. No neumonia class was obtained by ramdomly subsampling the dataset in order to avoid unbalanced classes.

** Training Dataset:** 

The training dataset was configured using a stratification strategy for the Pneumonia class using the subsampled dataset. It contains the 80% percent of the examples. A unique label for each disease was added to each example.

** Validation Dataset:** 

Validation dataset was also configured using a stratification strategy. It contains the 20% of the original dataset. 


### 5. Ground Truth

The labels were generated by automatic text analysis of radiologists's reports using a Natural Language Processing (NLP) algorithm. The NLP algorithm was reported to reach over 90% accuracy. NLP analysis is fast and cost effective but it has inherent issues which might affect the accuracy of the algorithm and limit its clinical applications under the given accuracy.

**Ground Truth Acquisition Methodology:**

The labels were generated by automatic text analysis of radiologists's reports using a Natural Language Processing (NLP) algorithm. The NLP algorithm was reported to reach over 90% accuracy. NLP analysis is fast and cost effective but it has inherent issues which might affect the accuracy of the algorithm and limit its clinical applications under the given accuracy.

Our recommendations for a ground truth acquisition methodology are:

- Improve labels by obtain biopsy data for all the validation samples
- Create a voting system among different radiologists (at least 3) and choose the majority label. 
- Create a bouding box indicating the areas of interest determined by the radiologist.
